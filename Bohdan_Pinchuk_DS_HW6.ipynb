```python
import numpy as np
import pandas as pd
import jinja2


class Reporter:
    """
    Class for collecting and printing data in tabular form
    """

    def __init__(self, attribute_col: str = "Attribute", result_col: str = "Result"):
        self._border_sign = "#"
        self._hor_split_sign = "~"
        self._ver_split_sign = "|"
        self._tolerance = 4
        self._max_len_lf: int = 0
        self._max_len_rt: int = 0
        self._data_list: list[tuple[str, str]] = []
        self.attribute_col = attribute_col
        self.result_col = result_col

    @property
    def max_len_lf(self) -> int:
        """Maximum left column width (the 1st column)"""
        return self._max_len_lf

    @property
    def max_len_rt(self) -> int:
        """Maximum right column width (the 2nd column)"""
        return self._max_len_rt

    @property
    def data_list(self) -> list[tuple[str, str]]:
        """The list of data to be printed"""
        return self._data_list

    @property
    def border_sign(self) -> str:
        """Character used for the outside table border"""
        return self._border_sign

    @border_sign.setter
    def border_sign(self, value: str) -> None:
        self._border_sign = value

    @property
    def hor_split_sign(self) -> str:
        """Character used for horizontal splitter in tabular form"""
        return self._hor_split_sign

    @hor_split_sign.setter
    def hor_split_sign(self, value: str) -> None:
        self._hor_split_sign = value

    @property
    def ver_split_sign(self) -> str:
        """Character used for vertical splitter in tabular form"""
        return self._ver_split_sign

    @ver_split_sign.setter
    def ver_split_sign(self, value: str) -> None:
        self._ver_split_sign = value

    @property
    def tolerance(self) -> int:
        """Number of digits after the floating point"""
        return self._tolerance

    @tolerance.setter
    def tolerance(self, value: int) -> None:
        self._tolerance = value

    @property
    def attribute_col(self) -> str:
        """Name of the attribute (the 1st column)"""
        return self._attribute_col

    @attribute_col.setter
    def attribute_col(self, value: str) -> None:
        array_str_key = value.splitlines()
        for str_key in array_str_key:
            self._max_len_lf = max(self.max_len_lf, len(str_key))
        self._attribute_col = value

    @property
    def result_col(self) -> str:
        """Name of the result (the 2nd column)"""
        return self._result_col

    @result_col.setter
    def result_col(self, value: str) -> None:
        array_str_value = value.splitlines()
        for str_value in array_str_value:
            self._max_len_rt = max(self.max_len_rt, len(str_value))
        self._result_col = value

    def format_value(self, value: float) -> str:
        """
        Format value to string
        :param value: input value
        :return: string of the value
        """
        return f"{value:.{self.tolerance}f}"

    def format_matrix(self, matrix: np.ndarray) -> str:
        """
        Formatting matrix into string
        :param matrix: input matrix
        :return: string of the matrix
        """
        return np.array2string(matrix, precision=self.tolerance)

    def add_item(self, key: str, value: str) -> None:
        """
        Form the report data that consist multi strings
        :param key: attribute
        :param value: result
        """
        array_str_key = key.splitlines()
        array_str_value = value.splitlines()

        for str_key in array_str_key:
            self._max_len_lf = max(self.max_len_lf, len(str_key))

        for str_value in array_str_value:
            self._max_len_rt = max(self.max_len_rt, len(str_value))

        self._data_list.append((key, value))

    def _print_line_splitter(self, sign: str = "-") -> None:
        """
        Print line splitter
        """
        print(sign * (self.max_len_lf + self.max_len_rt + 5))

    def _print_multi_string(self, row: tuple[str, str]) -> None:
        """
        Print the multi string
        :param row: tuple of data
        """
        array_str_key = row[0].splitlines()
        array_str_value = row[1].splitlines()
        num_str = max(len(array_str_key), len(array_str_value))

        for index in range(num_str):
            if index < len(array_str_key):
                left_str = f"{f"{array_str_key[index]}":<{self.max_len_lf}}"
            else:
                left_str = f"{f" ":<{self.max_len_lf}}"

            if index < len(array_str_value):
                right_str = f"{f"{array_str_value[index]:<{self.max_len_rt}}"}"
            else:
                right_str = f"{f" ":<{self.max_len_rt}}"

            print(f" {left_str} {self.ver_split_sign} {right_str}")

    def refresh_table_config(self) -> None:
        """
        Refresh table configuration, like line lengths
        :return:
        """
        self._max_len_lf = 0
        self._max_len_rt = 0

        self.attribute_col = self.attribute_col
        self.result_col = self.result_col

        temp_data_list = self.data_list
        self._data_list = []
        for temp_data in temp_data_list:
            self.add_item(temp_data[0], temp_data[1])

    def print_report(self) -> None:
        """
        Print report data
        """
        self._print_line_splitter(self.border_sign)
        self._print_multi_string((self.attribute_col, self.result_col))
        self._print_line_splitter(self.border_sign)
        for item_data in self.data_list[0:-1]:
            self._print_multi_string(item_data)
            self._print_line_splitter("~")
        self._print_multi_string(self.data_list[-1])
        self._print_line_splitter(self.border_sign)

    def print_refreshed_report(self) -> None:
        """
        Print report data with a refreshing table
        """
        self.refresh_table_config()
        self.print_report()

    def print_pd_report(self) -> None:
        """
        Print report data using pandas dataframe
        """
        df_data_list = pd.DataFrame(self.data_list, columns=[self.attribute_col, self.result_col])
        pd.set_option('display.max_colwidth', None)
        df_data_list = (df_data_list.style.hide(axis='index')
        .set_properties(**{
            'text-align': 'left',
            'white-space': 'pre-wrap',
            'border': '1px solid lightgrey'
        }).set_table_styles([  # type: ignore
            {"selector": "th", "props": [("text-align", "left")]}
        ]))
        display(df_data_list)
```

## Завдання 1. A/B тест для пропорцій

**Сценарій**:
Ви працюєте у мобільному додатку для замовлення їжі. Поточна конверсія покупки на головній сторінці (A) = 8%. Ви хочете протестувати нову версію головної сторінки (B). Очікуваний uplift = 9% → MDE = 1%.

**Завдання**:
1. Розрахуйте необхідний розмір вибірки на групу для тесту при α = 0.05 та power = 0.8.
2. Симулюйте A/B тест у Python (генерація випадкових даних для групи A та B).
3. Обчисліть p-value, абсолютну різницю та відносний uplift.
4. Інтерпретуйте результат: чи відкидаємо H0? Який бізнес-ефект?

**Підказка**: використайте `statsmodels.stats.proportion` та `proportions_ztest`.

```python
#### Variant №1 ####

import math
import numpy as np
import pandas as pd
from statsmodels.stats.power import NormalIndPower
from statsmodels.stats.proportion import proportion_effectsize
from statsmodels.stats.proportion import proportions_ztest

# Input data
p_A, p_B = 0.08, 0.09
alpha, power = 0.05, 0.8
n_simulations = 1000

# Solution
np.random.seed(42)
mde = p_B - p_A

# 1. Розрахунок вибірки
effect_size = proportion_effectsize(p_A, p_B)

analysis = NormalIndPower()

sample_size = analysis.solve_power(
    effect_size=effect_size,
    power=power,
    alpha=alpha,
    ratio=1.0
)
sample_size = math.ceil(sample_size)

# 2. Симуляція A/B тесту
group_A = np.random.binomial(1, p_A, sample_size)
group_B = np.random.binomial(1, p_B, sample_size)

successes = [group_A.sum(), group_B.sum()]
observations = [sample_size, sample_size]

# 3. Розрахунок розміру ефекту
# Z-test
z_stat, p_value = proportions_ztest(successes, observations)

p_conclusion = "значущий,\nнеобхідно перейти на нову сторінку" if p_value < 0.05 else "не значущий,\nне потрібно переходити на нову сторінку"

absolute_diff = group_B.mean() - group_A.mean()
relative_uplift = absolute_diff / group_A.mean()

# 4. Перевірка емпіричної потужності
significant_results = 0

for _ in range(n_simulations):
    sim_A = np.random.binomial(1, p_A, sample_size)
    sim_B = np.random.binomial(1, p_B, sample_size)
    sim_successes = [sim_A.sum(), sim_B.sum()]
    p_val = proportions_ztest(sim_successes, observations)[1]
    if p_val < alpha:
        significant_results += 1

empirical_power = significant_results / n_simulations

# Print results
df = pd.DataFrame({"group A": pd.Series(group_A), "group B": pd.Series(group_B)})
display(df)
rp = Reporter()
rp.tolerance = 6
rp.add_item("Конверсія покупки\nсторінка A", f"{p_A:.2%}")
rp.add_item("Конверсія покупки\nсторінка B", f"{p_B:.2%}")
rp.add_item("Uplift, MDE", f"{mde:.2%}")
rp.add_item("Рівень значущості", str(alpha))
rp.add_item("Потужність", str(power))
rp.add_item("Необхідний розмір вибірки на групу", f"{sample_size:,d}")
rp.add_item("Z-statistic", rp.format_value(float(z_stat)))
rp.add_item("p-value", rp.format_value(float(p_value)))
rp.add_item("Абсолютна різниця", rp.format_value(float(absolute_diff)))
rp.add_item("Абсолютна різниця, %", f"{absolute_diff:.2%}")
rp.add_item("Відносний uplift, %", f"{relative_uplift:.2%}")
rp.add_item("Гіпотеза H₀\n(відсутність ефекту, різниці або зв'язку)",
            "нова версія сторінки не вплинула на приріст покупок")
rp.add_item("Результат впливу на покупки\nпри впроваджені нової версії сторінки", p_conclusion)
rp.add_item("Емпірична потужність, %", f"{empirical_power:.2%}")
# rp.print_refreshed_report()
rp.print_pd_report()
```

## Завдання 2: A/B тест для continuous метрики

**Сценарій**:
Ви працюєте у SaaS-платформі для онлайн-курсів. Поточний середній дохід на користувача (ARPU) = 150$\$$. Ви тестуєте нову платіжну форму (B), очікуючи підвищення до 155$\$$ → MDE = 5$\$$. Стандартне відхилення доходу σ ≈ 30$\$$.

**Завдання**:
1. Розрахуйте необхідний розмір вибірки на групу для тесту при α = 0.05 та power = 0.8.
2. Симулюйте A/B тест у Python, згенеруйте дані для груп A та B.
3. Виконайте Welch t-test і обчисліть Cohen’s d.
4. Інтерпретуйте результат: чи можна сказати, що зміни статистично значущі? Який бізнес-ефект для 5 000 користувачів?

**Підказка**: використайте `scipy.stats.ttest_ind` та формулу для Cohen’s d:
$$
d=\frac{\overline{x}_B-\overline{x}_A}{\sqrt{\frac{s_A^2+s_B^2}{2}}}
$$

```python
#### Variant №1 ####

import math
import numpy as np
import pandas as pd
from scipy import stats
from statsmodels.stats.power import TTestIndPower

# Input data
mu_A, mu_B = 150, 155
sigma = 30
alpha, power = 0.05, 0.8
n_simulations = 1000
n_users = 5000

# Solution
mde = mu_B - mu_A
np.random.seed(42)

# 1. Розрахунок вибірки
effect_size = (mu_B - mu_A) / sigma

analysis = TTestIndPower()

sample_size = analysis.solve_power(
    effect_size=effect_size,
    power=power,
    alpha=alpha,
    ratio=1.0,
    alternative='two-sided'
)
sample_size = math.ceil(sample_size)

# 2. Симуляція A/B тесту (округлюємо до центів)
group_A = np.random.normal(mu_A, sigma, sample_size).round(2)
group_B = np.random.normal(mu_B, sigma, sample_size).round(2)

# 3. Розрахунок розміру ефекту
# Welch t-test (дисперсії можуть відрізнятися)
t_stat, p_value = stats.ttest_ind(group_B, group_A, equal_var=False)

p_conclusion = "значущий,\nнеобхідно перейти на нову платіжну форму" if p_value < 0.05 else "не значущий,\nне потрібно переходити на нову платіжну форму"

# Cohen's d
var_A = np.var(group_A, ddof=1)
var_B = np.var(group_B, ddof=1)

# alternative
# pooled_std = np.sqrt((group_A.std()**2 + group_B.std()**2) / 2)

# застосуємо спрощення для формули використовуючи 𝜎² = var
pooled_std = math.sqrt((var_A + var_B) / 2)
cohens_d = (group_B.mean() - group_A.mean()) / pooled_std

absolute_diff = group_B.mean() - group_A.mean()
relative_uplift = absolute_diff / group_A.mean()

# 4. Перевірка емпіричної потужності
significant_results = 0

for _ in range(n_simulations):
    sim_A = np.random.normal(mu_A, sigma, sample_size).round(2)
    sim_B = np.random.normal(mu_B, sigma, sample_size).round(2)
    p_val = stats.ttest_ind(sim_B, sim_A, equal_var=False)[1]
    if p_val < alpha:
        significant_results += 1

empirical_power = significant_results / n_simulations

# 5. Бізнес ефект
business_profit = absolute_diff.round(2) * n_users

# Print results
df = pd.DataFrame({"group A": pd.Series(group_A), "group B": pd.Series(group_B)})
display(df)
rp = Reporter()
rp.tolerance = 6
rp.add_item("Конверсія середнього доходу\nплатіжна форма A", f"${mu_A:.2f}")
rp.add_item("Конверсія середнього доходу\nплатіжна форма B", f"${mu_B:.2f}")
rp.add_item("Uplift, MDE", f"${mde:.2f}")
rp.add_item("Рівень значущості", str(alpha))
rp.add_item("Потужність", str(power))
rp.add_item("CKO, 𝜎", f"${sigma:.2f}")
rp.add_item("Необхідний розмір вибірки на групу", f"{sample_size:,d}")
rp.add_item("T-statistic", rp.format_value(float(t_stat)))
rp.add_item("p-value", rp.format_value(float(p_value)))
rp.add_item("Cohen's d", rp.format_value(float(cohens_d)))
rp.add_item("Середня різниця", f"${absolute_diff.round(2):.2f}")
rp.add_item("Відносний uplift, %", f"{relative_uplift:.2%}")
rp.add_item("Гіпотеза H₀\n(відсутність ефекту, різниці або зв'язку)",
            "нова платіжна форма не вплинула на приріст доходу")
rp.add_item("Результат впливу на дохід\nпри впроваджені нової платіжної форми", p_conclusion)
rp.add_item("Емпірична потужність, %", f"{empirical_power:.2%}")
rp.add_item(f"Додатковий дохід\n(із {n_users:,d} користувачів)", f"${business_profit:,.2f}")
# rp.print_refreshed_report()
rp.print_pd_report()
```
